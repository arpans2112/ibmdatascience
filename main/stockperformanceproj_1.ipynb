#### Extracting Stock Data Using Web Scraping

Not all stock data required for this project is available via a library. In this lab, you will use web scraping to obtain financial data. You will extract historical stock data from a web page using beautiful soup.

```python
# !mamba install bs4==4.10.0 -y
# !mamba install html5lib==1.1 -y 
# !pip install lxml==4.6.4
# !pip install mamba
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html"
```

```python
data  = requests.get(url).text
# print(data)
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
table_data = soup.find_all('tr')
print(table_data[0].prettify())
```

```python
row_data = table_data[0].find_all('span')
```

```python
#  create dataframe using the given columns. 
# Table Keys
netflix_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])
table_columns = {}
span_keys = table_data[0].find_all('span')
for i, span_key in enumerate(span_keys):
    table_columns[i] = span_key.text
print(table_columns)  

# Table data
for i,tablerow in enumerate(table_data):
    if i != 0:
        if len(table_data)-1 != i: 
            df_row_dict = {}
            row_data = tablerow.find_all('span')        
            for j,span_data in enumerate(row_data):
                df_row_dict[table_columns.get(j)] = span_data.text
            netflix_data = netflix_data.append(df_row_dict,ignore_index=True)

        
netflix_data['Close'] = netflix_data['Close*']
del(netflix_data['Close*'])
del(netflix_data['Adj Close**'])

# Final dataset. 
netflix_data.head()
    
```
